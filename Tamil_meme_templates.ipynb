```python
import os
import asyncio
from telethon import TelegramClient
from telethon.tl.types import InputMessagesFilterPhotos
 
# Telegram API credentials
api_id = '20244886'  # Replace with your API ID
api_hash = '2eaf972b7929683ae8cd14d893bdc06c'  # Replace with your API Hash
channel_username = 'tamiltemp'  # Replace with the public channel username
 
# Set up the download folder
download_folder = r'C:\Users\RamCharan\Downloads\tamil_templates'
if not os.path.exists(download_folder):
    os.makedirs(download_folder)
 
# Function to download all photos (memes)
async def download_memes():
    client = TelegramClient('anon', api_id, api_hash)
    await client.start()
 
    # Fetch the messages from the channel
    async for message in client.iter_messages(channel_username, filter=InputMessagesFilterPhotos):
        if message.photo:
            # Download the photo (meme)
            file_path = await client.download_media(message.photo, download_folder)
            print(f'Downloaded {file_path}')
 
    await client.disconnect()
 
# If running in an environment with an event loop (like Jupyter or PyCharm), use this:
if __name__ == "__main__":
    if not asyncio.get_event_loop().is_running():
        asyncio.run(download_memes())  # Works if no loop is running
    else:
        # If a loop is already running, just use `await`
        await download_memes()
```
