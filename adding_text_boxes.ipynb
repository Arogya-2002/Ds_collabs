```python
import cv2

# Global variables
input_text = [""]  # Use a list to store the text entered by the user
point = [-1, -1]    # Use a list to store the coordinates of the mouse click

def display_image_with_text(image, input_text, point):
    font, text_size, color, text_thickness = "","","",""
    
    def on_mouse_click(event, x, y, flags, param):
        nonlocal point
        nonlocal input_text ,font, text_size, color, text_thickness
        if event == cv2.EVENT_LBUTTONUP:
            point[0] = x
            point[1] = y
            input_text[0] = input("Enter text: ")  # Prompt user to enter text

            font_menu = """
                        Select Font:
                        1. cv2.FONT_HERSHEY_SIMPLEX
                        2. cv2.FONT_HERSHEY_PLAIN
                        3. cv2.FONT_HERSHEY_DUPLEX
                        4. cv2.FONT_HERSHEY_COMPLEX
                        5. cv2.FONT_HERSHEY_TRIPLEX
                        6. cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
                        """
            font_choice = int(input(font_menu + "Enter font choice (1-6): "))

            color_menu = """
                        Select Color:
                        1. White
                        2. Red
                        3. Green
                        4. Blue
                        5. Yellow
                        6. Custom (BGR format e.g., 255,0,0 for Red)
                        """
            color_choice = int(input(color_menu + "Enter color choice (1-6): "))
            if color_choice == 6:
                custom_color = input("Enter BGR color (e.g., 255,0,0 for Red): ")
                color = tuple(map(int, custom_color.split(',')))
            else:
                color_map = {1: (255, 255, 255), 2: (0, 0, 255), 3: (0, 255, 0), 4: (255, 0, 0), 5: (0, 255, 255)}
                color = color_map.get(color_choice)

            text_size = int(input("Enter text size (e.g., 1-5): "))
            text_thickness = int(input("Enter text thickness (e.g., 1-5): "))

            font_map = {
                1: cv2.FONT_HERSHEY_SIMPLEX,
                2: cv2.FONT_HERSHEY_PLAIN,
                3: cv2.FONT_HERSHEY_DUPLEX,
                4: cv2.FONT_HERSHEY_COMPLEX,
                5: cv2.FONT_HERSHEY_TRIPLEX,
                6: cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
            }
            font = font_map.get(font_choice)

    # Resize the image to desired dimensions
    desired_width = 500  # New width
    aspect_ratio = image.shape[1] / image.shape[0]
    desired_height = int(desired_width / aspect_ratio)
    resized_image = cv2.resize(image, (desired_width, desired_height))

    # Create a window and set the mouse callback function
    cv2.namedWindow("Image")
    cv2.setMouseCallback("Image", on_mouse_click)
    img_with_text = resized_image.copy()

    while True:
        if point != [-1, -1]:
            cv2.putText(img_with_text, input_text[0], tuple(point), font, text_size, color, text_thickness)

        cv2.imshow("Image", img_with_text)
        key = cv2.waitKey(1) & 0xFF
        if key == 27:  # Check if ESC key is pressed
            break
        elif key == ord('s'):  # Press 's' to save the image
            save_path = input("Enter the full file path to save the image (e.g., 'output.jpg'):")

            # Validate the file format
            if save_path.endswith(".jpg") or save_path.endswith(".png") or save_path.endswith(".bmp"):
                cv2.imwrite(save_path, img_with_text)  # Save the image
                print(f"Image saved at {save_path}")
            else:
                print("Invalid file format. Please use .jpg, .png, or .bmp.")

    cv2.destroyAllWindows()


def main():
    # Load an image (replace 'image.jpg' with your image path)
    image_path = r"C:\Users\SESPL\Desktop\litzchillData\telugu_templates\tests.jpg"
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found!")
        return

    # Display the resized image and allow user to interactively add text
    display_image_with_text(image, input_text, point)


if __name__ == "__main__":
    main()
```
