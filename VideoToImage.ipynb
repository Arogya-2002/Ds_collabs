```python
import cv2

video_path = 'Downloads/document_2024-11-20_09-56-00.mp4'
cap = cv2.VideoCapture(video_path)

frame_rate = cap.get(cv2.CAP_PROP_FPS)  # Get the frame rate of the video
count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Save a frame every 1 second
    if count % int(frame_rate) == 0:
        frame_filename = f'frame_{count // int(frame_rate)}.jpg'
        cv2.imwrite(frame_filename, frame)

    count += 2

cap.release()
```

```python
# video to images

import os
import cv2
 
def extract_images_from_video(video_path, output_dir):
    """
    Extract images from a video and save them in the output directory.
 
    :param video_path: Path to the input video file
    :param output_dir: Path to the output directory where images will be saved
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
 
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
 
    # Check if the video file was opened successfully
    if not video_capture.isOpened():
        print(f"Error: Could not open video file {video_path}")
        return
 
    # Get the total number of frames in the video
    total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))
 
    # Extract images at a fixed interval (every 10 frames in this example)
    frame_interval = 25
    for frame_num in range(0, total_frames, frame_interval):
        # Set the frame position
        video_capture.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
 
        # Read the frame
        ret, frame = video_capture.read()
 
        # Check if the frame was read successfully
        if not ret:
            print(f"Error: Could not read frame {frame_num} from video {video_path}")
            break
 
        # Save the frame as an image
        image_path = os.path.join(output_dir, f"{os.path.basename(video_path)}_frame{frame_num}.jpg")
        cv2.imwrite(image_path, frame)
 
    # Release the video capture object
    video_capture.release()
 
def extract_images_from_videos(input_dir, output_dir):
    """
    Extract images from all videos in the input directory and save them in the output directory.
 
    :param input_dir: Path to the input directory containing videos
    :param output_dir: Path to the output directory where images will be saved
    """
    # Get the list of video files in the input directory
    video_files = [f for f in os.listdir(input_dir) if os.path.isfile(os.path.join(input_dir, f)) and f.endswith((".mp4", ".avi", ".mov"))]
 
    # Extract images from each video file
    for video_file in video_files:
        video_path = os.path.join(input_dir, video_file)
        extract_images_from_video(video_path, output_dir)
 
if __name__ == "__main__":
    # Set the input and output directories
    input_dir = r"C:\Users\RamCharan\Downloads\televideos"
    output_dir = r"C:\Users\RamCharan\Downloads\VideoToImage_25f_2"
 
    # Extract images from all videos in the input directory
    extract_images_from_videos(input_dir, output_dir)
```
